```python

# Program to plot 2-D Heat map
# using matplotlib.pyplot.imshow() method
import numpy as np
import matplotlib.pyplot as plt
  
data = np.random.random((150 , 150))/10
plt.imshow( data , cmap = 'autumn' , interpolation = 'nearest' )
  
plt.title( "Relative error through space" )
norm = matplotlib.colors.Normalize(vmin=np.min(data), vmax=np.max(data))
m = cm.ScalarMappable(cmap=cm.autumn, norm=norm)
m.set_array([])
plt.colorbar(m, shrink = 0.5, anchor = (0.5, 0.5))
plt.show()
```
